```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

## PCK 2D and PCK 3D

```python
opt_2d = np.load("weights/optimized_keypoints/err_2d.npy")
marker_2d = np.load("weights/marker/err_2d.npy")
joints_2d = np.load("weights/joints/err_2d.npy")
DREAM_2d = np.load("weights/DREAM/err_2d.npy")
all_2d = np.load("weights/all_kp/err_2d.npy")

opt_3d = np.load("weights/optimized_keypoints/err_3d.npy")
marker_3d = np.load("weights/marker/err_3d.npy")
joints_3d = np.load("weights/joints/err_3d.npy")
DREAM_3d = np.load("weights/DREAM/err_3d.npy")
all_3d = np.load("weights/all_kp/err_3d.npy")


opt_3d = opt_3d * 1000
marker_3d = marker_3d * 1000
joints_3d = joints_3d * 1000
DREAM_3d = DREAM_3d*1000
all_3d = all_3d * 1000
```

```python
opt_pck_2d = []
for i in range(200):
    num = np.sum(opt_2d < i)
    opt_pck_2d.append(num/100)
    
marker_pck_2d = []
for i in range(200):
    num = np.sum(marker_2d < i)
    marker_pck_2d.append(num/100)
    
joints_pck_2d = []
for i in range(200):
    num = np.sum(joints_2d < i)
    joints_pck_2d.append(num/100)
    
DREAM_pck_2d = []
for i in range(200):
    num = np.sum(DREAM_2d < i)
    DREAM_pck_2d.append(num/100)
    
    
all_pck_2d = []
for i in range(200):
    num = np.sum(all_2d < i)
    all_pck_2d.append(num/100)
    
    
opt_pck_3d = []
for i in range(400):
    num = np.sum(opt_3d < i)
    opt_pck_3d.append(num/100)

    
marker_pck_3d = []
for i in range(400):
    num = np.sum(marker_3d < i)
    marker_pck_3d.append(num/100)
    
joints_pck_3d = []
for i in range(400):
    num = np.sum(joints_3d < i)
    joints_pck_3d.append(num/100)
    
    
DREAM_pck_3d = []
for i in range(400):
    num = np.sum(DREAM_3d < i)
    DREAM_pck_3d.append(num/100)
    
    
all_pck_3d = []
for i in range(400):
    num = np.sum(all_3d < i)
    all_pck_3d.append(num/100)

```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.subplot(211)
ax.plot(opt_pck_2d,label = "Optimized Keypoints (0.755)", linewidth=3,ls='-')
ax.plot(joints_pck_2d,label = "Keypoint at Joints (0.668)", linewidth=3 ,ls=':')
ax.plot(all_pck_2d,label = "All Candidate Keypoints (0.573)", linewidth=3,ls=':')
ax.plot(marker_pck_2d,label = "Aruco Marker (0.571)", linewidth=3 ,ls='-.')
ax.plot(DREAM_pck_2d,label = "DREAM[6] (0.440)", linewidth=3 ,ls='--')
ax.set_xlim(0,200)

plt.tick_params(labelsize = 14)
plt.grid(True)
plt.legend( prop={'size': 14})
plt.xlabel("threshold (pixel)",fontsize=16)
plt.ylabel("PCK (2D)",fontsize=16)

ax1 = plt.subplot(212)
ax1.plot(opt_pck_3d,label = "Optimized Keypoints (0.657)", linewidth=3,ls='-')
ax1.plot(joints_pck_3d,label = "Keypoint at Joints (0.591)", linewidth=3 ,ls=':')
ax1.plot(all_pck_3d,label = "All Candidate Keypoints (0.515)", linewidth=3,ls=':')
ax1.plot(marker_pck_3d,label = "Aruco Marker (0.434)", linewidth=3 ,ls='-.')
ax1.plot(DREAM_pck_3d,label = "DREAM[6] (0.406)", linewidth=3 ,ls='--')
#ax.set_xscale('log')
ax1.set_xlim(0,400)


plt.tick_params(labelsize = 14)
plt.grid(True)
plt.legend( prop={'size': 14})
plt.xlabel("threshold (mm)",fontsize=16)
plt.ylabel("PCK (3D)",fontsize=16)
#plt.savefig("pck_real.pdf")
```
